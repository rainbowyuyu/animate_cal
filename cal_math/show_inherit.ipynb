```python
from manim import *
from yty_manim.squ_tex import *
from yty_manim.matrix_yty import *
```

```python
%%manim -qh -v WARNING ShowBase
class ShowBase(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        t = Text("俞天翼").set_color(BLUE).scale(3)
        s = Square(fill_opacity=0.5).set_color(GREEN).scale(3)
        v = VGroup(s,t).arrange(buff=2)
        self.add(v)
```

```python
%%manim -qh -v WARNING ShowSquTex
type_dict = {
        "fill_opacity": 0,
        "stroke_opacity": 0,
    }

class ShowSquTex(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        text_t = "yty218"
        s_t = SquTex(text_t).set_color(BLUE).scale(1)
        s_t_c = s_t.copy()
        s_t_c.change_square(-1,-2,-3,color=PURPLE)
        s_t_c.change_text(0,1,2,color=TEAL)
        s_t_n = SquTex(text_t,**type_dict).set_color(GRAY).scale(1)
        s_t_v = VGroup(s_t,s_t_c,s_t_n).arrange(direction=DOWN,buff=0.5)
        self.add(s_t_v)
```

```python
%%manim -qh -v WARNING ShowMatrixCal
color_by_gradient_1 = [
    RED,
    ORANGE,
    YELLOW,
]
color_by_gradient_2 = [
    BLUE,
    TEAL,
    GREEN,
]

class ShowMatrixCal(Scene):
    def construct(self):
        self.camera.background_color = WHITE
        m1 = MatrixCal(matrix_3).set_color(BLACK)
        m2_o = MatrixCal(matrix_3,brackets_pair=['(',')']).set_color(BLACK)
        m2 = m2_o.neg_with_brackets().set_color(BLACK)
        m3 = MatrixCal(matrix_2t3).set_color(BLACK)
        m3_c = m3.get_column(0)
        m3_c.set_color_by_gradient(color_by_gradient_1)
        m4 = MatrixCal(matrix_3t2).set_color(BLACK)
        m4_r = m4.get_row(1)
        m4_r.set_color_by_gradient(color_by_gradient_2)
        sbg = Square(fill_opacity=0.5,stroke_opacity=0).set_color(BLUE)
        sbg.scale(0.6)
        m2[0][0].shift(UP*0.5)
        m2[0][1].set_color(RED)
        m2[0][2].scale(1.5).set_color(GREEN)
        m2[2][2].rotate(PI/2)
        m_v_1 = VGroup(m1,m2).arrange(buff=1)
        m_v_2 = VGroup(m3,m4).arrange(buff=1)
        m_v = VGroup(m_v_1,m_v_2).arrange(buff=0.5,direction=DOWN)
        sur = SurroundingRectangle(m2[2][0],buff=0)
        sbg.move_to(m2[2][1])
        self.add(m_v,sur,sbg)
```
